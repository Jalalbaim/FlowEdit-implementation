```python
df_agg = pd.read_csv("outputs_fig7/metrics_agg.csv")
df_agg
```

```python
# plot by method and link the points of the same method
import matplotlib.pyplot as plt

methods = df_agg['method'].unique()
for method in methods:
    subset = df_agg[df_agg['method'] == method]
    plt.plot(subset['clip_t_mean'], subset['lpips_mean'], marker='o', label=method)

plt.xlabel('Clip')
plt.ylabel('LPIPS')
plt.title('LPIPS by Clip')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```
